# **subplot / subplots 를 이용한 여러개의 차트 그리기**

```python
import matplotlib.pyplot as plt
```

\- 문제 : data1(x : 1, 2, 3; y : 1, 2, 3) 과 data2(x : 1, 2, 3; y : 1, 100, 200) 을 그래프로 출력하시오.

```python
x1 = [1, 2, 3]
y1 = [1, 2, 3]
x2 = [1, 2, 3]
y2 = [1, 100, 200]
```

* subplot을 이용한 해결

subplot(행의 수, 열의 수, 해당 그래프가 그려질 위치)

```python
plt.subplot(1, 2, 1) #1행 2열 표에 1,1에 그리기기
plt.plot(x1, y1)
plt.title('data1')

plt.subplot(1, 2, 2) #1행 2열 표에 1,2에 그리기기
plt.plot(x2, y2)
plt.title('data2')
```

```python
plt.subplot(2, 1, 1) #2행 1열 표에 1,1에 그리기기
plt.plot(x1, y1)
plt.title('data1')

plt.subplot(2, 1, 2) # 2행 1열 표에 2,1에 그리기기
plt.plot(x2, y2)
plt.title('data2')
```

\- tight_layout 을 이용하면 레이아웃이 자동으로 설정됨

```python
plt.subplot(2, 1, 1)
plt.plot(x1, y1)
plt.title('data1')

plt.subplot(2, 1, 2)
plt.plot(x2, y2)
plt.title('data2')
plt.tight_layout() # 글자가 겹치지 않게 
```

* subplots 를 이용한 해결

\- 문제 : data1(x : 1, 2, 3; y : 1, 2, 3) 과 data2(x : 1, 2, 3; y : 1, 50, 200) 을 그래프로 출력하시오.

```python
x1 = [1, 2, 3]
y1 = [1, 2, 3]
x2 = [1, 2, 3]
y2 = [1, 50, 200]
```

```python
fig, axe1 = plt.subplots()
#한개 표에 두개 그리기기
axe2 = axe1.twinx() #복사사
axe1.plot(x1, y1, color = 'red', label = 'data1')
axe2.plot(x2, y2, color = 'blue', label = 'data2')
#라벨 붙이기기
axe1.set_xlabel('x', fontsize = 15)
axe1.set_ylabel('y1', fontsize = 15)
axe2.set_ylabel('y2', fontsize = 15)
```

* y 축을 두 개 가진 차트에서 범례 표시하기

```python
fig, axe1 = plt.subplots()

axe2 = axe1.twinx()
chart1 = axe1.plot(x1, y1, color = 'red')
chart2 = axe2.plot(x2, y2, color = 'blue')

axe1.set_xlabel('x', fontsize = 15)
axe1.set_ylabel('y1', fontsize = 15)
axe2.set_ylabel('y2', fontsize = 15)
# chart 변수 만들어서 두개 chart를 더해 범례 표시시
chart = chart1 + chart2
axe1.legend(chart, ['data1', 'data2'])
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family']=['NanumGothic', 'sans-serif']
plt.rcParams['axes.unicode_minus'] = False

import seaborn as sns
import pandas as pd
```

\- 문제 : 다음은 2021년 영화별 관람자 수다. 이를 막대그래프로 표현하시오

```python
movie_title = ['크루엘라', '극장판 귀멸의 칼날: 무한열차편', '학교 가는 길']
audience = [664308, 2099131, 20067]

data = {'영화제목' : movie_title, '누적관객' : audience}
df = pd.DataFrame(data)
```

```python
df
```

```python
sns.barplot(data = df, x = '영화제목', y = '누적관객') #막대그래프프
```

\- 차트 크기 변경

```python
sns.set(rc={'figure.figsize':(10, 5)}) #10:5 비율로 세팅팅
```

```python
sns.barplot(data = df, x = '영화제목', y = '누적관객')
```

\- 누적관객수 별로 그리기

```python
sns.barplot(data = df, x = '영화제목', y = '누적관객',
             order = df.sort_values('누적관객').영화제목)
```

\- ascending 을 이용한 내림차순 정렬

```python
sns.barplot(data = df, x = '영화제목', y = '누적관객',
             order = df.sort_values('누적관객', ascending = False).영화제목)
```

\- 관객 수 포맷 변환

```python
chart = sns.barplot(data = df, x = '영화제목', y = '누적관객',
             order = df.sort_values('누적관객', ascending = False).영화제목)

ylabels = ['{:,.0f}'.format(i) + '만 명' for i in chart.get_yticks() / 10000]
chart.set_yticklabels(ylabels)
```

\- 경고문 제외

```python
import matplotlib.ticker as mticker

chart = sns.barplot(data = df, x = '영화제목', y = '누적관객',
             order = df.sort_values('누적관객', ascending = False).영화제목)

ticks_labels = chart.get_yticks().tolist()
chart.yaxis.set_major_locator(mticker.FixedLocator(ticks_labels))
chart.set_yticklabels(['{:,.0f}'.format(i / 10000) + '만 명' for i in ticks_labels])
```

\- 가로로 그리기

```python
chart = sns.barplot(data = df, x = '누적관객', y = '영화제목',
             order = df.sort_values('누적관객', ascending = False).영화제목)

xlabels = ['{:,.0f}'.format(i) + '만 명' for i in chart.get_xticks() / 10000]
chart.set_xticklabels(xlabels)
```

\- 차트 색상 변경

```python
chart = sns.barplot(data = df, x = '누적관객', y = '영화제목',
             order = df.sort_values('누적관객', ascending = False).영화제목,
             color = 'blue')

xlabels = ['{:,.0f}'.format(i) + '만 명' for i in chart.get_xticks() / 10000]
chart.set_xticklabels(xlabels)
```

\- 제목 추가, 폰트 사이즈 변경

```python
chart = sns.barplot(data = df, x = '누적관객', y = '영화제목',
             order = df.sort_values('누적관객', ascending = False).영화제목,
             color = 'blue')

xlabels = ['{:,.0f}'.format(i) + '만 명' for i in chart.get_xticks() / 10000]
chart.set_xticklabels(xlabels)

plt.xlabel('누적관객', fontsize = 15)
plt.ylabel('영화제목', fontsize = 15)
plt.title('영화 별 누적관객수', fontsize = 20)
```

# **scatterplot을 이용한 산점도 그리기**

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize':(10, 5)})
import pandas as pd
```

\- 문제 : iris 데이터 셋을 이용하여 sepal_length, sepal_width 별 산점도를 작성하시오.

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df.isnull().sum() #결측값 확인인
```

```python
sns.scatterplot(data = df, x = 'sepal_width', y = 'sepal_length')
```

\- hue 를 이용한 종 분류

```python
sns.scatterplot(data = df, x = 'sepal_width', y = 'sepal_length', hue = 'species')
```

\- 추가 문제 : petal_length 의 값을 3 미만, 5 미만, 5 이상으로 분류하여 표시하여라.

```python
df['petal_length'].unique()
```

```python
def function1(x):
    if x < 3:
        return 's'
    elif x < 5:
        return 'm'
    else:
        return 'l'
```

```python
df['petal_length2'] = df['petal_length'].apply(function1)
```

```python
df.head()
```

```python
sns.scatterplot(data = df, x = 'sepal_width', y = 'sepal_length', hue = 'species', style = 'petal_length2')
```

\- 점의 크기로 분류

```python
def function2(x):
    if x < 3:
        return 3
    elif x < 5:
        return 5
    else:
        return 7
```

```python
df['petal_length3'] = df['petal_length'].apply(function2)
```

```python
df.head()
```

```python
sns.scatterplot(data = df, x = 'sepal_width', y = 'sepal_length', hue = 'species', size = 'petal_length3')
```

\- relplot 을 이용하면 카테고리 별로 따로 그릴 수 있음

```python
sns.relplot(data = df, x = 'sepal_width', y = 'sepal_length', col = 'species')
```

\- lmplot 을 사용하면 회귀선을 그릴 수 있음

```python
sns.lmplot(data = df, x = 'sepal_width', y = 'sepal_length', hue = 'species')
```

# **displot / histplot 을 이용한 히스토그램 그리기**

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize':(15, 5)})
import pandas as pd
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

## distplot

```python
sns.distplot(df['sepal_length']) #구간자동으로 나눠서 표시
```

\- 구간 개수 설정

```python
sns.distplot(df['sepal_length'], bins = 15)
```

## histplot

```python
sns.histplot(df['sepal_length'])
```

\- 구간 개수 설정

```python
sns.histplot(df['sepal_length'], bins = 20) #구간 개수 설정정
```

\- kde 추가

```python
sns.histplot(df['sepal_length'], bins = 20, kde = True)
```

\- count -> density

```python
sns.histplot(df['sepal_length'], bins = 20, kde = True, stat = 'density')
```

\- 카테고리 별 분류

```python
sns.histplot(df['sepal_length'], bins = 20, kde = True, stat = 'density', hue = 'species')
```

```python
sns.histplot(data = df, x = 'sepal_length', bins = 20, kde = True, stat = 'density', hue = 'species')
```

\- x 축, y 축 변경

```python
sns.histplot(data = df, y = 'sepal_length', bins = 20, kde = True, stat = 'density', hue = 'species')
```

\- 차원 추가

```python
sns.histplot(data = df, y = 'sepal_length', x = 'sepal_width')
```

\- 칼라바 추가

```python
sns.histplot(data = df, y = 'sepal_length', x = 'sepal_width', cbar = True)
```

## displot

```python
sns.displot(df['sepal_length'])
```

```python
sns.displot(data = df, x = 'sepal_length')
```

\- 그림 크기 변경

```python
sns.displot(data = df, x = 'sepal_length', height = 5, aspect = 3) # width= height * aspect
```

\- histgram 과 한 옵션 사용(1)

```python
a = sns.displot(data = df, y = 'sepal_length', bins = 20, kde = True, stat = 'density', hue = 'species')
```

\- histgram 과 한 옵션 사용(2)

```python
sns.displot(data = df, y = 'sepal_length', x = 'sepal_width', cbar = True)
```

## jointplot

\- 기본 사용법

```python
sns.jointplot(data = df,  x = 'sepal_width', y = 'sepal_length')
```

\- histplot 의 형식으로 변경

```python
sns.jointplot(data = df,  x = 'sepal_width', y = 'sepal_length', kind = 'hist', cbar = True) #kind 를 hist 형식으로로
```
